```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("D:/data handling/road_safety_data.csv")

# Display basic information about the dataset
print(df.info())

# Display the first few rows of the dataset
print(df.head())
```

```python
# Data Visualization
# Analyzing road safety based on Casualty_Severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Casualty_Severity', data=df)
plt.title('Casualty Severity Distribution')
plt.xlabel('Casualty Severity')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd

# Prompt user for accident severity
severity_input = input("Enter the severity of the accident (Normal/Severe): ").strip().lower()

# Create DataFrame with recommendations based on severity input
if severity_input == "normal":
    recommendations = {
        "Precautions": [
            "Always obey traffic rules and regulations.",
            "Avoid distractions while driving, such as texting or talking on the phone.",
            "Ensure proper maintenance of vehicles to prevent mechanical failures.",
            "Use seat belts and child restraints properly.",
            "Be cautious and attentive, especially in adverse weather conditions.",
            "Encourage pedestrian and cyclist safety through awareness programs.",
            "Advocate for improved infrastructure and road safety measures."
        ]
    }
elif severity_input == "severe":
    recommendations = {
        "Precautions": [
            "Avoid speeding and reckless driving behaviors.",
            "Maintain a safe distance from other vehicles.",
            "Avoid driving under the influence of alcohol or drugs.",
            "Ensure proper vehicle maintenance and tire condition.",
            "Use advanced safety features available in vehicles.",
            "Take additional precautions in adverse weather conditions.",
            "Advocate for stricter enforcement of traffic laws and regulations."
        ]
    }
else:
    print("Invalid input. Please enter 'Normal' or 'Severe'.")

# Create DataFrame
df_recommendations = pd.DataFrame(recommendations)

# Export DataFrame to CSV file
if severity_input in ["normal", "severe"]:
    file_name = f"accident_precautions_{severity_input}.csv"
    df_recommendations.to_csv(file_name, index=False)
    print(f"Recommendations for {severity_input} accidents saved to {file_name}.")
```
